```python
# Dropping the rows with missing values in the column GDP per capita will effect particular countries. 
# Therfore, I will imput the values by avg the GDP per capita per this country from non missing years.
```

```python
happy.groupby('Country name').agg({'Log GDP per capita': ['mean']})
```

```python
for i in happy.groupby('Country name').agg({'Log GDP per capita': ['mean']}).index:#range(0,happy.groupby('Country name').agg({'Log GDP per capita': ['mean']}).shape[0],1):
    for j in happy[(happy['Country name']==i) & (happy['Log GDP per capita'].isnull()==True)].index: #print(happy[(happy['Country name']==i) & (happy['Log GDP per capita'].isnull()==True)]['Log GDP per capita'])
    #print(happy.groupby('Country name').agg({'Log GDP per capita': ['mean']}).loc[i])
        #print(i, j)
        #print(happy['Log GDP per capita'][j])
        #print(happy[(happy['Country name']==i) & (happy['Log GDP per capita'].isnull()==True)]['Log GDP per capita'][j])
        happy['Log GDP per capita'][j] = happy.groupby('Country name').agg({'Log GDP per capita': ['mean']}).loc[i].values
        #print(happy.groupby('Country name').agg({'Log GDP per capita': ['mean']}).loc[i].values)
```
